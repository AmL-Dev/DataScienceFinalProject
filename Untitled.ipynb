```python
import pandas as pd
import numpy as np
from collections import defaultdict
```

```python
df_user = pd.read_csv("data/user_book_ratings.csv")
df_book = pd.read_csv("data/books_metadata.csv")
```

```python
df_user.head(2000)
```

```python
df_book
```

```python
df_users = df_user.head(2000).groupby('user_id')['book_id'].apply(list).reset_index(name='movies_list')
df_ratings = df_user.head(2000).groupby('user_id')['rating'].apply(list).reset_index(name='ratings_list')
df_ratings
```

```python

def list_converter(number_list): 
    author_list = []
    for i in number_list:
        name = df_book.loc[df_book["book_id"]==i,"author"].to_string(index=False)
        author_list.append(name)
    return author_list
```

```python
df_users["authors"]= df_users.apply(lambda row : list_converter(row["movies_list"]),axis=1)
df_users["ratings"]= df_ratings["ratings_list"]
df_users
```

```python
def list_combiner(authors_list, ratings_list):
    # Group the items by author.
    value_lists = defaultdict(list)
    for author, rating in zip(authors_list, ratings_list):
        value_lists[author].append(rating)

    # Take the average of each author list.
    result = {}
    for author, rating in value_lists.items():
        result[author] = sum(ratings_list) / float(len(ratings_list))
    return result
     
df_users["author_rating"]= df_users.apply(lambda row : list_combiner(row["authors"],row["ratings"]),axis=1)

    
# averages(["john","danny","john","john"],[0,5,5,5])
```
